# Canny Edge Detection
Canny edge detection goes through the following steps
1. Canny use a low pass gaussian filter to cancel out noise. 
2. finds the strength and direction of edges using Sobel filters
3. Applies non-maximum supresion to isolate the strongest edges and thin them to one pixel wide lines
4. uses the hysterises to isolate the best edges

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
image = cv2.imread('Brain_image.jpeg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
#Convert image to grayscale for processing
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap = 'gray')
```

## Impelement Canny edge detection

```python
#Using wide and tight thresholds for hysterises
wide = cv2.Canny(gray,30,100)
tight = cv2.Canny(gray, 200,240)
#Diaplay the image
f, (ax1,ax2) = plt.subplots(1,2,figsize = (20,10))
ax1.set_title('wide')
ax1.imshow(wide, cmap = 'gray')

ax2.set_title('tight')
ax2.imshow(tight,cmap='gray')
```

```python
#Examin different threshold values
wide = cv2.Canny(gray,90,100)
tight = cv2.Canny(gray, 120,240)
#Diaplay the image
f, (ax1,ax2) = plt.subplots(1,2,figsize = (20,10))
ax1.set_title('wide')
ax1.imshow(wide, cmap = 'gray')

ax2.set_title('tight')
ax2.imshow(tight,cmap='gray')
```
