# K-means Clustering
K-means clustering for image segmentation

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

image = cv2.imread('fruits.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (20,10))
plt.imshow(image)
```

## Reshaping image and converting to float type

```python
#Reshape image into a 2D array of pixel and 3 color values
image_pixels = image.reshape((-1,3))
#convert to float type
pixel_values = np.float32(image_pixels)
```

## K-means Clustering

```python
#Define the K-mean clustering parameters 
#Criteria: the number of maximum iternations for faster convergance
#Select K to perform k-means clustering

criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
k = 6 
retval,labels, centers = cv2.kmeans(pixel_values, k, None, criteria, 100, cv2.KMEANS_RANDOM_CENTERS)
#Convert data into 8-bit values
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]
#Reshape data into original image dimention
segmented_image = segmented_data.reshape((image.shape))
labels_reshape = labels.reshape(image.shape[0], image.shape[1])
plt.figure(figsize = (20,10))
plt.imshow(segmented_image)
```
