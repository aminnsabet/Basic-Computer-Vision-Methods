# HSV Color Space

## Import resourrces and dispaly image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#read in the image
image = cv2.imread('Ballon.jpeg')
#change color to RGB for BGR
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

## visualizing the color channel

```python
#RGB channels
r = image[:,:,0]
g = image[:,:,1]
b = image[:,:,2]
f, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,10))
ax1.set_title('Red')
ax1.imshow(r,cmap = 'gray')

ax2.set_title('Green')
ax2.imshow(g,cmap='gray')

ax3.set_title('Blue')
ax3.imshow(b,cmap='gray')
```

## Visualize HSV color channels

```python
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
#HSV channel
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

f, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,10))
ax1.set_title('Hue')
ax1.imshow(h,cmap = 'gray')

ax2.set_title('Saturation')
ax2.imshow(s,cmap='gray')

ax3.set_title('Value')
ax3.imshow(v,cmap='gray')
```

## Define pink and hue selection threshold

```python
# Define our color selection criteria in HSV values
lower_hue = np.array([160,0,0])
upper_hue = np.array([180,255,255])
```

```python
# Define our color selection criteria in RGB values
lower_pink = np.array([180,0,100])
upper_pink = np.array([255,255,230])
```

## Mask image

```python
#Define mask area in rgb space
mask_rgb = cv2.inRange(image,lower_pink,upper_pink)
#mask the image
masked_image = np.copy(image)
masked_image[mask_rgb==0] = [0,0,0]
#visualize the image
plt.imshow(masked_image)
```

```python
#Define mask in hsv space
mask_hsv = cv2.inRange(hsv,lower_hue,upper_hue)
#mask the image
masked_image = np.copy(image)
masked_image[mask_hsv==0] = [0,0,0]
#visualize the image
plt.imshow(masked_image)
```
