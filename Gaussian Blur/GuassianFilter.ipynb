# Gaussian Blur for medical images

one of most frequently used filtered in computer vision. Its a weighted average which gives the most weightvto centre pixel but also take into account surronding pixels

1/16*[[1,2,1],
      [2,4,2],
      [1,2,1]]

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

#Read Image in
image = cv2.imread('Brain_image.jpeg')
#make a copy of image
image_copy = np.copy(image)
#change the color channels for BGR to RGB
image_copy =  cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
#visualize image
plt.imshow(image_copy)
```

## Gaussian blur the image

```python
#convert to grayscale for filtering
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
#OpenCv provide the Gussian filter
gray_blur = cv2.GaussianBlur(gray,(9,9),0)
f,(ax1,ax2) = plt.subplots(1,2,figsize = (20,10))
#Visualizeing iamges
ax1.set_title('original image')
ax1.imshow(gray,cmap = 'gray')
ax2.set_title('blurred image')
ax2.imshow(gray_blur,cmap = 'gray')
```

## Test performance with a high pass filter

High pass filters are used to sharpen an image
Enhance high-frequency part of image
Low pass filters smooth/Blur an image and block high frequency of image
Low pass filters like Gaussian filter use to remove noise

```python
#3*3 high pass filter

sobel_x = np.array([[-1,0,1],
                    [-2,0,2],
                    [-1,0,1]])

# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel) 
filtered_gray_blur = cv2.filter2D(gray_blur,-1,sobel_x)
filtered_gray = cv2.filter2D(gray,-1,sobel_x)

#visualizing detected edges
f,(ax1,ax2) = plt.subplots(1,2,figsize = (20,10))

ax1.set_title("With Gaussian blur")
ax1.imshow(filtered_gray_blur, cmap ='gray')
ax2.set_title('Original photo')
ax2.imshow(filtered_gray, cmap ='gray')

```
